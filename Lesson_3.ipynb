1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
import numpy as np
import math
```

```python
salaries = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]
M = sum(salaries)/len(salaries)
print(f'Среднее арифметическое выборки равно {M}')
print(np.mean(salaries))
```

```python
salaries_variance = sum([(item - M)**2 for item in salaries])/(len(salaries)-1)
salaries_variance_offset = sum([(item - M)**2 for item in salaries])/(len(salaries))
print(f'Смещенная оценка дисперсии равна - {round(salaries_variance_offset, 1)}')
print(f'Несмещенная оценка дисперсии равна - {round(salaries_variance, 1)}')
sigma = math.sqrt(salaries_variance)
print((f'Среднее квадратическое отклонение выборки равно {round(sigma, 1)}'))
```

2. В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

```python
def combinations (n, k):
    if k != 0:
        return math.factorial(n)/math.factorial(k)/math.factorial(n-k)
    else:
        return 1
```

```python
'''
три белых мяча могут получиться в трех случаях: 
1. из первого ящика достали 0 белых - 2 черных, а из второго 3 белых - 1 черный 
2. из первого ящика достали 1 белый - 1 черный, а из второго 2 белых - 2 черных 
3. из первого ящика достали 2 белых - 0 черных, а из второго 1 белый - 3 черных
'''

p1 = combinations(5, 0)*combinations(3, 2)/combinations(8, 2)*combinations(5, 3)*combinations(7, 1)/combinations(12,4)
p2 = combinations(5, 1)*combinations(3, 1)/combinations(8, 2)*combinations(5, 2)*combinations(7, 2)/combinations(12,4)
p3 = combinations(5, 2)*combinations(3, 0)/combinations(8, 2)*combinations(5, 1)*combinations(7, 3)/combinations(12,4)

print(f'Вероятность того, что три мяча белые равна - {p1 + p2 + p3}')
```

4. В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

```python
p_A = 1/4 #вероятность, что студент учится на факультете А 
p_B = 1/4 #вероятность, что студент учится на факультете B 
p_C = 2/4 #вероятность, что студент учится на факультете B 

p_A_complete = 0.8
p_B_complete = 0.7 
p_C_complete = 0.9
P_full = p_A*p_A_complete + p_B*p_B_complete + p_C*p_C_complete

p1 = p_A*p_A_complete/P_full
print(f'Вероятность, что сдавший студент учится на факультете A - {p1}')

p2 = p_B*p_B_complete/P_full
print(f'Вероятность, что сдавший студент учится на факультете B - {p2}')

p3 = p_C*p_C_complete/P_full
print(f'Вероятность, что сдавший студент учится на факультете C - {p3}')
```
